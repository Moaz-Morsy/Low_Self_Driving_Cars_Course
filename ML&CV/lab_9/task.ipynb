```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
colored_circles = cv2.imread('Files/circles.png')
gray_circles = cv2.cvtColor(colored_circles, cv2.COLOR_BGR2GRAY)
gray_circles = cv2.medianBlur(gray_circles,5)
colored_capture = cv2.imread('Files/Capture.PNG')
gray_capture = cv2.cvtColor(colored_capture, cv2.COLOR_BGR2GRAY)
gray_capture = cv2.medianBlur(gray_capture,5)
```

```python
# plt.figure(figsize=(15, 10))
plt.imshow(colored_circles[:,:,::-1]);
```

```python
# plt.figure(figsize=(25, 10))
plt.imshow(colored_capture[:,:,::-1]);
```

```python
def nothing(x):
    pass
```

```python
cv2.namedWindow('output', cv2.WINDOW_NORMAL)
cv2.createTrackbar('chose_image', 'output', 0, 1, nothing)
cv2.createTrackbar('x', 'output', 1, 100, nothing)
cv2.createTrackbar('y', 'output', 20, 100, nothing)
cv2.createTrackbar('param1', 'output', 50, 100, nothing)
cv2.createTrackbar('param2', 'output', 30, 100, nothing)
cv2.createTrackbar('minR', 'output', 20, 100, nothing)
cv2.createTrackbar('maxR', 'output', 50, 100, nothing)

while 1:
    chose_image = cv2.getTrackbarPos('chose_image','output')
    x = cv2.getTrackbarPos('x', 'output')
    y = cv2.getTrackbarPos('y', 'output')
    param1 = cv2.getTrackbarPos('param1', 'output')
    param2 = cv2.getTrackbarPos('param2', 'output')
    minR = cv2.getTrackbarPos('minR', 'output')
    maxR = cv2.getTrackbarPos('maxR', 'output')
    
    if chose_image == 0:
        cimg = colored_circles.copy()
        gray = gray_circles
    elif chose_image == 1:
        cimg = colored_capture.copy()
        gray = gray_capture
    #cimg = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
    #cimg = colored_img.copy()

    if x < 1:
        x = 1
    if y < 1:
        y = 1
    if param1 < 1:
        param1 = 1
    if param2 < 1:
        param2 = 1
    if minR > maxR:
        minR = maxR - 1
    circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, x, y,
                            param1=param1,param2=param2,minRadius=minR,maxRadius=maxR)

    #print(circles.shape)
    if circles is not None:
        circles = np.uint16(np.around(circles))

        for num,i in enumerate(circles[0,:]):
            # draw the outer circle
            cv2.circle(cimg,(i[0],i[1]),i[2],(0,255,0),2)
            # draw the center of the circle
            font = cv2.FONT_HERSHEY_SIMPLEX
            cv2.putText(cimg, f"{num}", (i[0]-5,i[1]+5), font, 0.5, (0, 0, 0), 1, cv2.LINE_AA)
            #cv2.circle(cimg,(i[0],i[1]),2,(0,0,255),3)

    cv2.imshow('output', cimg)
    
    k = cv2.waitKey(1) & 0xFF
    if k == 27:
        break
    elif k == ord('s'):
        if chose_image == 0:
            cv2.imwrite(f"Files/circles_edges.jpg",cimg)
        elif chose_image == 1:
            cv2.imwrite(f"Files/capture_edges.jpg",cimg)

cv2.destroyAllWindows()
```
