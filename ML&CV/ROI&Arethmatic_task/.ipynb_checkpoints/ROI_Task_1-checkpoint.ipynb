```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import imutils
```

```python
img = cv2.imread("Files/carkyo.jpg")

plt.imshow(img[:,:,::-1]);
```

```python
copy = img.copy()

plt.imshow(copy[:,:,::-1]);
```

```python
logo = cv2.imread("Files/tekomoro.png")

plt.imshow(logo[:,:,::-1]);
```

```python
print(copy.shape)
print(logo.shape)
```

```python
# logo_resize = cv2.resize(logo,(round(logo.shape[1]*(150/logo.shape[1])),round(logo.shape[0]*(150/logo.shape[1]))))
logo_resize = cv2.resize(logo,None,fx=(150/logo.shape[1]),fy=(150/logo.shape[1]),interpolation = cv2.INTER_CUBIC)
```

```python
logo_resize.shape
```

```python
logo_resize = imutils.resize(logo, width=150)
```

```python
logo_resize.shape
```

```python
plt.imshow(logo_resize[:,:,::-1]);
```

```python
gray = cv2.cvtColor(logo_resize, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray');
```

```python
ret, mask = cv2.threshold(gray, 50, 255, cv2.THRESH_BINARY_INV)
plt.imshow(mask, cmap='gray');
```

```python
mask_inv = cv2.bitwise_not(mask)
plt.imshow(mask_inv, cmap='gray');
```

```python
height_margin = copy.shape[0]-logo_resize.shape[0]-10
width_margin = copy.shape[1]-logo_resize.shape[1]-10
```

```python
roi = copy[height_margin:-10,width_margin:-10]
plt.imshow(roi[:,:,::-1]);
```

```python
img_bg = cv2.bitwise_and(roi, roi, mask=mask)
plt.imshow(img_bg[:,:,::-1]);
```

```python
img_fg = cv2.bitwise_and(logo_resize, logo_resize, mask=mask_inv)
plt.imshow(img_fg[:,:,::-1]);
```

```python
dst = cv2.add(img_bg, img_fg)
plt.imshow(dst[:,:,::-1]);
```

```python
copy[height_margin:-10,width_margin:-10] = dst
plt.imshow(copy[:,:,::-1]);
```

```python
cv2.imwrite(f"Files/copy.jpg",copy)
```
