```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('Files/Affine/perspective01.png')
```

```python
plt.imshow(img[:,:,::-1]);
```

```python
copy = img.copy()
```

```python
copy.shape
```

```python
cols = copy.shape[1]
rows = copy.shape[0]
```

```python
cv2.namedWindow('Image', cv2.WINDOW_NORMAL)

while 1:
    pts1 = np.float32(
        [
            [325, 186], [0, 253],
            [958, 325], [578, 185]
        ]
    )

    pts2 = np.float32(
        [
            [325, 0], [325, 539],
            [578, 539], [578, 0]
        ]
    )

    M = cv2.getPerspectiveTransform(pts1 ,pts2)
    dst = cv2.warpPerspective(copy, M, (cols,rows))

    
        
#     for point in pts2:
#         dst = cv2.circle(copy, (int(point[0]), int(point[1])), 1, (0, 0, 255), 2)

#     for point in pts1:
#         dst = cv2.circle(copy, (int(point[0]), int(point[1])), 1, (0, 255, 0), 2)
#     out = dst

    cv2.imshow('Image', dst)
    k = cv2.waitKey(1)
    if k == ord('q'):
        break
    elif k == ord('s'):
            cv2.imwrite(f"task_2.jpg",dst)

cv2.destroyAllWindows()
```
