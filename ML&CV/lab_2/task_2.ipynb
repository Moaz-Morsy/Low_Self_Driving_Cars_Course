```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.colors
```

```python
img = cv2.imread('Files/001.jpg')
```

```python
plt.imshow(img);
```

```python
YCrCb = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
YCrCb_fix = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
```

```python
plt.imshow(YCrCb);
```

```python
YCrCb[:,:,0] = YCrCb[:,:,0]*1.25 # Y channel
YCrCb_fix[:,:,0] = cv2.multiply(YCrCb_fix[:,:,0],1.25) # Y channel
```

```python
plt.imshow(YCrCb);
```

```python
plt.imshow(YCrCb_fix);
```

```python
bgr = cv2.cvtColor(YCrCb, cv2.COLOR_YCrCb2BGR)
bgr_fix = cv2.cvtColor(YCrCb_fix, cv2.COLOR_YCrCb2BGR)
```

```python
plt.imshow(bgr[:,:,::-1]);
```

```python
plt.imshow(bgr_fix[:,:,::-1]);
```

```python
plt.imshow(img[:,:,::-1]);
```

#### What is the effect of each change?
- Change Y-channel will lead to dark spots in image (problem)

#### What problem did you face and how you fixed it?
- Problem: Color Overflow will lead to dark spots in image
- Solution: use cv2.multiply and cv2.add will perform a color saturation and dark spots will be adjusted to its normal values
- Fixed image will be more brighter

```python
cv2.imwrite(f"Files/bgr_task_2.jpg",bgr_fix)
```
